```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Распределение Пуассона
### генерируем выбору из 1000 значений, gamma(n+1) это раширенние факториал на все числа

```python
x = np.arange(0,20,0.1)
def f(x):
    return np.math.pow(6,x)/np.math.gamma(x+1)*np.exp(-6)
     
y=x
n=0;
for v in x:
    y[n]=f(x[n])
    n=n+1
res = np.zeros((20,1))
#print res
#print y

smp = np.arange(0,1000,1)
for n in np.arange(0,1000,1):
    en= True
    while en:
        s = np.random.randint(0,20)
        r= np.random.random()
        if (r<f(s)):
            res[s]+=1
            smp[n]=s
            en=False
```

### строим графики

```python
res=res/1000
        # res[v] = res[v] + np.random.choice(np.arange(0,20,1), 1)
res=  np.array(res.T)[0,:]
x = np.arange(0,20,0.1)
xi = np.arange(0,20,1.)
#print res ,xi, res.shape[0], xi.shape[0]

plt.plot(x, y, label='theoretical')
#plt.step(xi, res, label='experimental')
plt.hist(smp, bins=14,normed=True )
plt.legend(loc='upper right')
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

### генерируем k выборок объема 5, 20, 50 (1000 у меня комп долго считает)

```python
k=400
smp3 = np.zeros((k,3))
smp20 = np.zeros((k,20))
smp50 = np.zeros((k,50))

for i in np.arange(0,k,1):
    for j in np.arange(0,3,1):
        en= True
        while en:
            s = np.random.randint(0,20)
            r= np.random.random()
            if (r<f(s)):
                smp3[i,j]=s
                en=False
    for j in np.arange(0,20,1):
        en= True
        while en:
            s = np.random.randint(0,20)
            r= np.random.random()
            if (r<f(s)):
                smp20[i,j]=s
                en=False
    for j in np.arange(0,50,1):
        en= True
        while en:
            s = np.random.randint(0,20)
            r= np.random.random()
            if (r<f(s)):
                smp50[i,j]=s
                en=False
```

### Считаем средние

```python


av3 = np.arange(0,k,1.)
av20 = np.arange(0,k,1.)
av50 = np.arange(0,k,1.)
for  n in np.arange(0,k,1):
    av3[n]=np.average(smp3[n,:])
for  n in np.arange(0,k,1):
    av20[n]=np.average(smp20[n,:])
for  n in np.arange(0,k,1):
    av50[n]=np.average(smp50[n,:])
```

### рисуем гистограммы и графики нормального распределения
### для дисперсии нормального берем дисперсию пуассона (6) и делим на размер  выборки

```python

x=np.arange(0,14,0.1)
gr3 = sts.norm(6,np.math.sqrt(6/3.)).pdf(x)
gr20 = sts.norm(6,np.math.sqrt(6/20.)).pdf(x)
gr50 = sts.norm(6,np.math.sqrt(6/50.)).pdf(x)
#plt.step(xi, res, label='experimental')
plt.hist(av50, bins=8,normed=True )
plt.hist(av20, bins=8,normed=True )
plt.hist(av3, bins=8,normed=True )

plt.plot(x, gr50, label='theoretical for 50')
plt.plot(x,gr20,label='theoretical for 20')
plt.plot(x,gr3,label='theoretical for 3')
plt.legend(loc='upper right')
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

### Видно что при больших n дисперсия понижается, как и говорит ЦПТ
