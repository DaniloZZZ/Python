```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
# Исследуем распределение хи-квадрат с параметром 3. Делаем выборку объема 1000, строим гистограмму и график теоретической плотности вероятности.

k = 3

rv_chi2 = sts.chi2(k)
sample = rv_chi2.rvs(1000)
plt.hist(sample, normed = True, bins = 40)

x = np.linspace(0,20,100)
pdf = rv_chi2.pdf(x)
plt.plot(x, pdf, label='theoretical PDF')
plt.legend(loc='upper right')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# В этом блоке вычисляем выборочное среднее для 1000 выборок объема 5, строим гистограмму его эмпирического распределения.

n = 5

# Вычисляем выборочное среднее для каждой из 1000 выборок
sample_de = []
for i in xrange(0, 1000):
    sample = rv_chi2.rvs(n)
    sample_e = 0
    for value in sample:
        sample_e = sample_e + value / n  # вычисляем выборочное среднее
    sample_de.append(sample_e)

plt.hist(sample_de, bins = 20, normed = True)

# строим график теоретического распределения. 
# Из табличных значений для распределения хи-квадрат с параметром k его матожидание равно k, дисперсия - 2k.
# По ЦПТ распределение выборочного среднего должно быть нормальным с параметрами E = k, D = 2k/n, где n - объем выборки.

norm_rv = sts.norm(k, math.sqrt(2. * k / n))

x = np.linspace(0,9,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label = 'theoretical pdf')
plt.legend(loc='upper right')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# В этом блоке вычисляем выборочное среднее для 1000 выборок объема 10, строим гистограмму его эмпирического распределения.

n = 10

# Вычисляем выборочное среднее для каждой из 1000 выборок
sample_de = []
for i in xrange(0, 1000):
    sample = rv_chi2.rvs(n)
    sample_e = 0
    for value in sample:
        sample_e = sample_e + value / n
    sample_de.append(sample_e)

plt.hist(sample_de, bins = 20, normed = True)

# строим график теоретического распределения. 
# Из табличных значений для распределения хи-квадрат с параметром k его матожидание равно k, дисперсия - 2k.
# По ЦПТ распределение выборочного среднего должно быть нормальным с параметрами E = k, D = 2k/n, где n - объем выборки.

norm_rv = sts.norm(k, math.sqrt(2. * 3 / n))
x = np.linspace(0, 6, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf, label = 'theoretical pdf')
plt.legend(loc='upper right')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# В этом блоке вычисляем выборочное среднее для 1000 выборок объема 50, строим гистограмму его эмпирического распределения.

n = 50

# Вычисляем выборочное среднее для каждой из 1000 выборок 
sample_de = []
for i in xrange(0, 1000):
    sample = rv_chi2.rvs(n)
    sample_e = 0
    for value in sample:
        sample_e = sample_e + value / n
    sample_de.append(sample_e)

plt.hist(sample_de, bins = 20, normed = True)

# строим график теоретического распределения. 
# Из табличных значений для распределения хи-квадрат с параметром k его матожидание равно k, дисперсия - 2k.
# По ЦПТ распределение выборочного среднего должно быть нормальным с параметрами E = k, D = 2k/n, где n - объем выборки.

norm_rv = sts.norm(k, math.sqrt(2. * k / n))
x = np.linspace(0, 4.5, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)
plt.plot(x, pdf, label = 'theoretical pdf')
plt.legend(loc='upper left')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
