```python
# импортируем необходимые библиотеки:
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline 
```

```python
# генерируем выборку объема 1000 из распределения Парето:
b = 3.0
pareto_rv = sts.pareto(b)
sample = pareto_rv.rvs(1000)
```

```python
# строим гисторграмму выборки объема N = 1000 и теоретическую плотность распределения (pdf) Парето:
plt.hist(sample, normed = 'True', label='sample')
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.title('Pareto')

x = np.linspace(sts.pareto.ppf(0.01, b), sts.pareto.ppf(0.99, b), 1000)
pdf = pareto_rv.pdf(x)
plt.plot(x, pdf, 'r-', lw = 2, label='theoretical pdf')
plt.legend()
```

```python
# вычисляем теоретическое среднее, дисперсию и СКО распределения Парето:
import math
Xm = 1.0
E = b*Xm/(b-1) # мат.ожидание
D = Xm/(b-1)**2*(b/(b-2)) # дисперсия
G = math.sqrt(D) # СКО
print 'E =', E
print 'D =', D
print 'G =', G
```

```python
# генерируем 1000 выборок объема N = 5, определяем выборочное среднее каждой выборки и записываем его в массив:
n5 = []
for i in range(1, 1001):
    sample5 = pareto_rv.rvs(5.)
    n5.append(sample5.mean())
      
# строим гисторграмму распределения выборочных средних при объеме выборки N = 5:
plt.hist(sample5, bins = 50, normed=True, label='sample')
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.title('N=5')

# строим плотность соответствующего нормального распределения:
norm_rv = sts.norm(E, G/5)
x = np.linspace(sts.pareto.ppf(0.01, b), sts.pareto.ppf(0.99, b), 1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, 'r-', lw = 2, label='normal')
plt.legend()

# вычисляем параметры нормального распределения, аппроксимирующего выборочные средние при различных N = 5:
print 'E =', norm_rv.mean()
print 'D =', norm_rv.var()
print 'G =', norm_rv.std()
```
